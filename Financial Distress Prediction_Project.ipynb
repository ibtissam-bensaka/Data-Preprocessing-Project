# Splitting the dataset into training and testing datasets

```python
from sklearn.model_selection import train_test_split # used for splitting training and testing data
```

```python
# Splitting the attributes into independent and dependent attributes
#X = df.iloc[:, :-1].values # attributes to determine dependent variable / Class
#Y = df.iloc[:, -1].values # dependent variable / Class
X = df.drop('Financial_Distress',axis=1)
y = df['Financial_Distress']
y = (y<-0.5)
y
```

```python
#Time = np.unique(df["x78"])
#print(Time.size)
# splitting the dataset into training set and test set
X_train, X_test, Y_train, Y_test = train_test_split(X, y, test_size=0.3, random_state=0)
print('Training records:',Y_train.size)
print('Test records:',Y_test.size)
```

#  feature scaling

```python
from sklearn.preprocessing import StandardScaler # used for feature scaling
#preprocessing.MinMaxScaler()
```

```python
# feature scaling
#We use feature scaling to convert different scales 
#to a standard scale to make it easier for Machine Learning algorithms.
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

# Multilayer Perceptron without dimension reduction

```python
from sklearn.neural_network import MLPClassifier
```

```python
# Initializing the multilayer perceptron
#multilayer perceptron with 1 hidden_layer
mlp1 = MLPClassifier(hidden_layer_sizes=(53),solver='adam',learning_rate_init= 0.01, max_iter=500)
# Train the model
mlp1.fit(X_train, Y_train)
```

```python
loss_values = mlp1.loss_curve_
plt.plot(loss_values)
plt.show()
```

```python
#multilayer perceptron with 2 hidden_layer
mlp2 = MLPClassifier(hidden_layer_sizes=(53,53),solver='adam',learning_rate_init= 0.01, max_iter=500)
# Train the model
mlp2.fit(X_train, Y_train)
```

```python
loss_values = mlp2.loss_curve_
plt.plot(loss_values)
plt.show()
```

```python
#multilayer perceptron with 3 hidden_layer
mlp3 = MLPClassifier(hidden_layer_sizes=(53,53,53),solver='adam',learning_rate_init= 0.01, max_iter=500)
# Train the model
mlp3.fit(X_train, Y_train)
```

```python
loss_values = mlp3.loss_curve_
plt.plot(loss_values)
plt.show()
```

# Predictions

```python
predictions1 = mlp1.predict(X_test)
predictions2 = mlp2.predict(X_test)
predictions3 = mlp3.predict(X_test)
```

# Evaluation

```python
from sklearn.metrics import classification_report,confusion_matrix
from sklearn.metrics import accuracy_score
```

```python
print("Mutilayers Perceptron with 1 layers")
print(confusion_matrix(Y_test,predictions1))
print("Mutilayers Perceptron with 2 layers")
print(confusion_matrix(Y_test,predictions2))
print("Mutilayers Perceptron with 3 layers")
print(confusion_matrix(Y_test,predictions3))
```

```python
print("==============Mutilayers Perceptron with 1 layers")
print('Accuracy' ,accuracy_score(Y_test, predictions1))
print(classification_report(Y_test,predictions1))
print("==============Mutilayers Perceptron with 2 layers")
print('Accuracy' , accuracy_score(Y_test, predictions2))
print(classification_report(Y_test,predictions2))
print("==============Mutilayers Perceptron with 3 layers")
print('Accuracy' , accuracy_score(Y_test, predictions3))
print(classification_report(Y_test,predictions3))
```

# PCA

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA()
pca
# Transformation with its own data
X_train = pca.fit_transform(X_train)
X_test = pca.transform(X_test)
pca_variance = pca.explained_variance_
pca_variance
plt.figure(figsize=(15, 6))
plt.bar(range(85), pca_variance, alpha=0.5, align='center', label='individual variance')
#plt.bar(np.arange(X_train.all()), pca_variance, alpha=0.5, align='center', label='individual variance')
plt.legend()
plt.ylabel('Variance ratio')
plt.xlabel('Principal components')
plt.show()
```

```python
pca = PCA(n_components=1)
X_train = pca.fit_transform(X_train)
X_test = pca.transform(X_test)
```

```python
#multilayer perceptron with 2 hidden_layer
mlp = MLPClassifier(hidden_layer_sizes=(53,53),solver='adam',learning_rate_init= 0.01, max_iter=500)
# Train the model
mlp.fit(X_train, Y_train)
```

```python
# Predicting the Test set results
y_pred = mlp.predict(X_test)
```

```python
#Evaluation
cm = confusion_matrix(Y_test, y_pred)
print(cm)
print('Accuracy' , accuracy_score(Y_test, y_pred))
```
